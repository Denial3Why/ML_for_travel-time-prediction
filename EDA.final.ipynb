```python
plt.figure(figsize=(10,6))
sns.heatmap(df3.corr(), cmap='coolwarm', annot=True)
sns.set(rc={'figure.facecolor':(0,0,0,0)})
# Mengatur ukuran font untuk label sumbu x dan y
plt.tick_params(axis='x', labelsize=16)  # Ukuran font untuk label sumbu x
plt.tick_params(axis='y', labelsize=16)  # Ukuran font untuk label sumbu y

plt.title('Correlation Heatmap ',
         fontsize=18)
```
